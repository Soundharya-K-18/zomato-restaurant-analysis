```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('zomato.csv', encoding='latin-1')
```

```python
print(df.shape)

print(df.dtypes)
```

```python
df.head()
```

```python
def clean_rate(val):
  try:
      return float(str(val).split('/')[0].strip())
  except:
      return np.nan
df['rate']=df['rate'].apply(clean_rate)
```

```python
df.head()
```

```python
rate=print(df['rate'].describe())
```

```python
df['approx_cost(for two people)']=(df['approx_cost(for two people)'].astype(str).str.replace(',','').str.strip())
```

```python

df.dtypes
```

```python
df.head()
```

```python
df['approx_cost(for two people)']=pd.to_numeric(df['approx_cost(for two people)'],errors='coerce')
df.dtypes
```

```python
df.rename(columns={'approx_cost(for two people)':'cost_for_two','listed_in(type)':'restaurant_type','listed_in(city)':'city'},inplace=True)
```

```python
df.head(2)
```

```python
cost=print(df['cost_for_two'].describe())
```

```python
df.dropna(subset=['name','city'],inplace=True)
```

```python
#print(df.isnull().sum())
```

```python
print(df.duplicated().sum())
```

```python
df.drop_duplicates(subset=['name','city'],inplace=True)
```

```python
df.shape
```

```python
df.to_csv('zomato_clean.csv',index=False)
```
