```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('zomato_clean.csv')
```

```python
sns.set_theme(style='whitegrid')
```

```python
#chart-1: Top areas by restaurant count
plt.figure(figsize=(10,6))
top_cities=df['city'].value_counts().head(10)
sns.barplot(x=top_cities.values,y=top_cities.index,palette='Blues_r')
plt.title('Top 10 Areas in Bangalore by restaurant count',fontsize=12)
plt.xlabel('Number of Restaurants')
plt.tight_layout()
#plt.savefig('charts/chart1_top_areas.png',dpi=150)
plt.show()          
```

```python
#chart-2: Rating distribution 
plt.figure(figsize=(10,5))
sns.histplot(df['rate'].dropna(),bins=30,color='steelblue',kde=True)
plt.title('Distribution of Restaurant Rating',fontsize=12)
plt.xlabel('Rating')
plt.tight_layout()
#plt.savefig('charts/chart2_rating_dist.png',dpi=150)
plt.show()
```

```python
#chart-3:Online order vs rating
plt.figure(figsize=(10,6))
sns.boxplot(x='online_order',y='rate',data=df)
plt.title('Rating:Online Order vs No Online Order',fontsize=12)
plt.tight_layout()
plt.show
```

```python
#chart-4:Cost Vs Rating Scatter
plt.figure(figsize=(10,6))
sns.scatterplot(x='cost_for_two',y='rate',data=df.sample(2000),alpha=0.4,color='coral')
plt.title('Cost for Two vs Rating',fontsize=12)
plt.xlabel('cost for Two ₹')
plt.ylabel('Rating')
plt.tight_layout()
plt.show()
```

```python
#Chart-5:Restaurant type distribution
plt.figure(figsize=(10,5))
type_counts=df['restaurant_type'].value_counts()
sns.barplot(x=type_counts.values,y=type_counts.index,palette='Greens_r')
plt.title('Restaurant Count by Type',fontsize=12)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
top_cuisines=(df.groupby('cuisines')['rate'].agg(['mean','count']).query('count>100').sort_values('mean',ascending=False).head(10))
sns.barplot(x=top_cuisines['mean'],y=top_cuisines.index,palette='Oranges_r')
plt.title('Top 10 Cuisines by Average Rating',fontsize=12)
plt.xlabel('Average Rating')
plt.tight_layout()
plt.show()
```
